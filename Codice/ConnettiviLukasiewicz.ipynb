```
def al(var1, var2):
  return min (var1, var2)

def ol(var1,var2):
  return max (var1, var2)

def fl(var1, var2):
  return min(1, 1 - var1 + var2)

def xl(var1,var2):
  return 1 - abs(var1 - var2)

def nl(var1):
  return 1 - (var1)


def call(formula):
 neg_pos = (False, True)  
 
 if(formula == 'nl'): 
   print(' _________________') 
   print('|      A     NOT  |')
   for A in neg_pos: 
    print(f' NOT {A} = {eval(formula+"(A)")}')
   print('|_________________|')
 if(formula == 'al'):
   print(' ___________________________')
   print('|  A         B        AND   |')
   for A in neg_pos:
     for B in neg_pos:
       print(f' {A} AND {B}  =  {eval(formula+"(A,B)")}  ')
   print('|___________________________|')
 if(formula == 'ol'):
   print(' ___________________________')
   print('|  A        B         OR    |')
   for A in neg_pos:
     for B in neg_pos:
       print(f' {A} OR {B}  =  {eval(formula+"(A,B)")}  ')
   print('|___________________________|')
 if(formula == 'fl'):
   print(' ___________________________')
   print('|  A        B         ->    |')
   for A in neg_pos:
     for B in neg_pos:
       print(f' {A} -> {B}  =  {eval(formula+"(A,B)")}  ')
   print('|___________________________|')
 if(formula == 'xl'):
   print(' ___________________________')
   print('|  A        B        XOR    |')
   for A in neg_pos:
     for B in neg_pos:
       print(f' {A} XOR {B}  =  {eval(formula+"(A,B)")}  ')
   print('|___________________________|')
```

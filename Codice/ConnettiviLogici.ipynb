```
#Funzioni della logica proposizionale

def a(var1,var2): #funzione and
  return var1 and var2

def o(var1,var2): #funzione or
  return var1 or var2

def f(ipotesi, tesi): # A e B funzione per chiamare il connettivo freccia
  return not ipotesi or tesi

def x(var1,var2): #funzione xor
  return (not var1 and var2) or (var1 and not var2)

def n(var1):
  return not var1


def call(formula): #funzione per chiamare i connettivi - è incasinata solo per fare qualche effetto grafico :(
 neg_pos = (False, True)  
 
 if(formula == 'n'): #ho messo tante if solo perchè volevo far apparire un risultato
   print(' _________________') #grafico differente e quindi la scritta NOT o AND o altro - so che non è il massimo
   print('|      A     NOT  |')
   for A in neg_pos: 
    print(f' NOT {A} = {eval(formula+"(A)")}')
   print('|_________________|')
 if(formula == 'a'):
   print(' ___________________________')
   print('|  A         B        AND   |')
   for A in neg_pos:
     for B in neg_pos:
       print(f' {A} AND {B}  =  {eval(formula+"(A,B)")}  ')
   print('|___________________________|')
 if(formula == 'o'):
   print(' ___________________________')
   print('|  A        B         OR    |')
   for A in neg_pos:
     for B in neg_pos:
       print(f' {A} OR {B}  =  {eval(formula+"(A,B)")}  ')
   print('|___________________________|')
 if(formula == 'f'):
   print(' ___________________________')
   print('|  A        B         ->    |')
   for A in neg_pos:
     for B in neg_pos:
       print(f' {A} -> {B}  =  {eval(formula+"(A,B)")}  ')
   print('|___________________________|')
 if(formula == 'x'):
   print(' ___________________________')
   print('|  A        B        XOR    |')
   for A in neg_pos:
     for B in neg_pos:
       print(f' {A} XOR {B}  =  {eval(formula+"(A,B)")}  ')
   print('|___________________________|')

```
